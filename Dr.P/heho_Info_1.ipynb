# 透過爬蟲蒐集資料

```python
#匯入會使用到的套件包
import urllib.request as req #抓取網頁原始碼(HTML)
import pandas as pd
import requests 
import time
from time import sleep
from bs4 import BeautifulSoup
```

```python
#抓取的網站
url = 'https://heho.com.tw/archives/category/health_pic/acupoint-exercise' 

# 建立 Request 物件，附加 Request Headers(因為連線被拒絕，必須看起來像是人類)
request = req.Request(url, headers = {
    "User-Agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'
})
with req.urlopen(request) as response:
    data = response.read().decode("utf-8")
    
res = requests.get(url)
res
```

```python
#以 BeautifulSoup 解析 HTML 程式碼
soup = BeautifulSoup(res.text, 'html.parser')
```

# 抓取各類文章標題、連結、圖片

```python
# 以 CSS 的 class 抓出各類文章標題
titles = soup.find_all("h5", class_="post-title is-large") 

for t in titles:
    #標題
    print("文章標題:", t.text)
    print()
```

```python
links = soup.find_all("a", class_="plain")

for l in links:
    href = l.get('href')
    if "heho-menu" in href:
        continue
    print("文章連結:", href)
    print()
```

```python
from IPython.display import Image

images = soup.find_all("img", class_="lazy-load attachment-medium size-medium wp-post-image")

for img in images:
    i = img.get('data-src')
    if i is not None:
        print ("文章縮圖:",i)
        print()
```

## 所以我決定先收集標題和圖片的資料就好，再蒐集連結的資料，再把兩個合在一起

```python
#先抓10頁的內容測試看看可不可以正確抓到資料

import bs4
max_pages = 10
for page in range(1, max_pages + 1):
    # 構造每一頁的URL
    page_url = f"{url}/page/{page}"  # 根據網站的實際URL結構進行調整
    
    # 發送HTTP請求獲取網頁內容
    response = requests.get(page_url)
    
    # 使用BeautifulSoup解析網頁內容
    root = bs4.BeautifulSoup(response.text, "html.parser")
    
    titles = root.find_all("h5", class_="post-title is-large")
    
    images = root.find_all("img", class_="lazy-load attachment-medium size-medium wp-post-image")

    for t, img  in zip(titles, images):

        # 標題
        print("標題:", t.text)


        # 圖片
        image = img.get('data-src')
        if image is not None:
            print("圖片:", image)

        print()
            
    print("-------------------第", page,"頁-------------------")
    print()
```

```python
#試著把抓下來的資料放進Dataframe裡以利後續資料整理

data1 = []

for page in range(1, max_pages + 1):
    # 構造每一頁的URL
    page_url = f"{url}/page/{page}"  # 根據網站的實際URL結構進行調整
    
    # 發送HTTP請求獲取網頁內容
    response = requests.get(page_url)
    
    # 使用BeautifulSoup解析網頁內容
    root = bs4.BeautifulSoup(response.text, "html.parser")
    
    titles = root.find_all("h5", class_="post-title is-large")
    
    images = root.find_all("img", class_="lazy-load attachment-medium size-medium wp-post-image")

    for t, img  in zip(titles, images):

        title = t.text

        image = img.get('data-src')

        data1.append([title, image])

df = pd.DataFrame(data1, columns=['Title', 'Image'])
df
```

## 爬取每一頁關於標題和圖片的資料，並放入Dataframe裡

```python
data2 = []
page = 1  # 起始頁數
has_next_page = True  # 是否還有下一頁

while has_next_page:
    # 構造每一頁的 URL
    page_url = f"{url}/page/{page}"  # 根據網站的實際 URL 結構進行調整
    
    # 發送 HTTP 請求獲取網頁內容
    response = requests.get(page_url)
    
    # 使用 BeautifulSoup 解析網頁內容
    soup = BeautifulSoup(response.text, "html.parser")
    
    # 爬取頁面的相關資料
    titles = soup.find_all("h5", class_="post-title is-large")

    images = soup.find_all("img", class_="lazy-load attachment-medium size-medium wp-post-image")
    
    # 在這裡處理資料或存儲到 DataFrame 中
    for t, img  in zip(titles, images):

        title = t.text

        image = img.get('data-src')

        data2.append([title, image])
    
    # 檢查是否還有下一頁
    next_button = soup.find("a", class_="next page-number")  # 根據網站的實際 HTML 結構進行調整
    has_next_page = next_button is not None
    
    page += 1  # 頁數加一

# 在這裡進行最終的資料處理或結果輸出
df = pd.DataFrame(data2, columns=['Title', 'Image'])
df
```

## 爬取每一頁關於文章連結的資料，並放入另一個Dataframe裡

```python
data_links = []
page = 1  # 起始頁數
has_next_page = True  # 是否還有下一頁

while has_next_page:
    # 構造每一頁的 URL
    page_url = f"{url}/page/{page}"  # 根據網站的實際 URL 結構進行調整
    
    # 發送 HTTP 請求獲取網頁內容
    response = requests.get(page_url)
    
    # 使用 BeautifulSoup 解析網頁內容
    soup = BeautifulSoup(response.text, "html.parser")
    
    # 爬取頁面的相關資料
    links = soup.find_all("a", class_="plain")
    
    for link in links:
        href = link.get('href')
        if "heho-menu" in href:
            continue

        data_links.append([href])

    # 檢查是否還有下一頁
    next_button = soup.find("a", class_="next page-number")  # 根據網站的實際 HTML 結構進行調整
    has_next_page = next_button is not None
    
    page += 1  # 頁數加一

# 在這裡進行最終的資料處理或結果輸出
df_links = pd.DataFrame(data_links, columns=['Link'])
df_links
```

```python
df_links.to_csv('heho_links_1.csv', index=False)
print("ok")
```
