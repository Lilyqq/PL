```python
#文字探勘
import jieba
import jieba.analyse
import pandas as pd
import numpy as np
from collections import Counter
```

```python
# 測試結巴功能
documents = ["我是師大Jisoo"]
# 精確模式
for sentence in documents:
    test = jieba.cut(sentence)
    print('/'.join(test))
```

```python
import csv
jieba.load_userdict('dict1.txt')
data1 = []

# 打開CSV文件
with open('heho.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    
    # 逐行讀取CSV數據
    for row in reader:
        title = row['Title']
        
        # 使用jieba進行分詞
        
        keywords = jieba.lcut(title, HMM=False)
       
        # 過濾出詞彙表中的詞
        with open('dict1.txt', 'r', encoding='utf-8') as dict_file:
            dict1 = dict_file.read().splitlines()  # 讀取詞彙表文件並按行分割為列表
        
        selected_keywords = [word for word in keywords if word in dict1]
        
        # 將標點符號去除
        selected_keywords = [word.replace('[','').replace(']','').replace(',', '') for word in selected_keywords]
        
        #print(selected_keywords)
        
        data1.append([selected_keywords])

tag_data = pd.DataFrame(data1, columns=['Tag'])
tag_data
```

```python
tag_data.to_csv('Tag.csv', index=False)
```

```python
data['Tag'] = tag_data['Tag']
data
```
